# Data Visualization

Importing requied libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing the data 

```python
marvel=pd.read_csv("marvel.csv") 
dc=pd.read_csv("dc.csv")
```

### 1. To visualize the number of each charcter type:

To know the number and types of ALIGN

```python
set(marvel['ALIGN'])  
```

```python
set(dc['ALIGN'])
```

```python
plt.figure(figsize=(8,5))
x=[0,2,4,6]
w=[0.9,2.9,4.9,6.9]
y=[len(marvel[marvel['ALIGN']=='Bad Characters']),len(marvel[marvel['ALIGN']=='Good Characters']),len(marvel[marvel['ALIGN']=='Neutral Characters']),len(marvel[marvel['ALIGN']=='Reformed Criminals'])]
z=[len(dc[dc['ALIGN']=='Bad Characters']),len(dc[dc['ALIGN']=='Good Characters']),len(dc[dc['ALIGN']=='Neutral Characters']),len(dc[dc['ALIGN']=='Reformed Criminals'])]
labels=['Bad Characters', 'Good Characters', 'Neutral Characters','Reformed Criminals']
plt.bar(x,y,color='y',width=0.9,label='Marvel')
plt.bar(w,z,color='b',width=0.9,label='DC')
plt.xticks(x,labels)
plt.xticks(w,labels)
plt.xlabel('Character type', fontsize=15)
plt.ylabel('Number of appearances',fontsize=15)
plt.legend()
plt.title('Number of characters in each character type',fontsize=18)
plt.show()
```

### 2.Plotting the appearances of the characters:

```python
plt.figure(figsize=(50,15))
sns.barplot(x=marvel['Year'].tolist(),y=marvel['APPEARANCES'].tolist())
plt.xticks(rotation=45)
plt.title('Marvel - Appearance of character in each year',fontsize=20)

plt.figure(figsize=(50,15))
sns.barplot(x=dc['YEAR'].tolist(),y=dc['APPEARANCES'].tolist())
plt.xticks(rotation=45)
plt.title('DC - Appearance of character in each year',fontsize=20)
```

### 3. Plotting overall gender evolvement in Marvel and DC according to the year:

```python
dc.groupby('YEAR')['SEX'].value_counts().unstack().plot(figsize = (16,6))
plt.title('DC - Gender Evolvement')

marvel.groupby('Year')['SEX'].value_counts().unstack().plot(figsize = (16,6))
plt.title('Marvel - Gender Evolvement')
```

### 4. Total number of male and female charcters

```python
A=['Male Characters','Female Characters']
B=[len(marvel[marvel['SEX']=='Male Characters']),len(marvel[marvel['SEX']=='Female Characters'])]
C=[len(dc[dc['SEX']=='Male Characters']),len(dc[dc['SEX']=='Female Characters'])]
f, ax = plt.subplots(figsize = (6,3))
sns.barplot(x=B,y=A,color='orange',alpha=0.5,label='Marvel')
sns.barplot(x=C,y=A,color='black',alpha=0.7,label='DC')

ax.legend() 
ax.set(xlabel='Number of Characters', ylabel='Sex',title = "Number of male and female characters")
```

### 5. Swarm plot of status of existence of characters by sex and year

```python
sns.swarmplot(x="SEX", y="Year",hue="ALIVE", data=marvel)
plt.show()
```

```python
sns.swarmplot(x="SEX", y="YEAR",hue="ALIVE", data=dc)
plt.show()
```
